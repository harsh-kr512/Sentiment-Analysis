```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
file = pd.read_excel('Input.xlsx')
file.head(5)
```

```python
urls=file['URL']
ids = file['URL_ID']
```

```python

def save_text_file(url, id):
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.content, 'html.parser')
        para = soup.find_all('p')
        paragraphs = [p.get_text() for p in para]

        paragraph = ' '.join(paragraphs)
        filename = str(id) + '.txt'

        with open(filename, 'w', encoding='utf-8') as f:
            f.write(paragraph)

        print(f"Text extracted from '{url}' and saved as '{filename}'")
    else:
        print(f"Failed to retrieve '{url}'. Status code: {response.status_code}")

# Iterate over URLs and IDs and save text files
for url, id in zip(urls, ids):
    save_text_file(url, id)
```
